```python
import mlxtend
import pandas as pd
import numpy as np
import plotly.express as px
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
#importation du dataset
df = pd.read_csv('Market_Basket_Optimisation.csv', header=None)
nombreDeColonne = len(df.columns)
nomDeColonne = [f'Colonne {i+1}' for i in range(nombreDeColonne)]
df.columns = nomDeColonne
df.head()
```

```python
df=df.loc[:,['Colonne 1','Colonne 2']]
fig = px.area(df, x = "Colonne 1", y = "Colonne 2", color = "Colonne 1")
fig.show()
```

```python
# convertion en np.array et suppression des valeus NaN
data = [[value for value in row if pd.notna(value)] for row in df.values]
print(data)
```

```python
# encodage en format bool
te=TransactionEncoder()
te_ary=te.fit(data).transform(data)
df=pd.DataFrame(te_ary, columns=te.columns_)  
df
```

```python
#Slectionnons des ensembles dlments avec un minimum de 10 % de prise en charge
apriori(df, min_support=0.1)
```

```python
frequent_itemsets=apriori(df, min_support=0.1, use_colnames=True)
frequent_itemsets
```
